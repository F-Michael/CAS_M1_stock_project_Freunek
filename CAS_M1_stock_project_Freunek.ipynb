```python
#Test der Normalverteilung:

print('p-value from the normality test on the setal widths and lengths (D-Agostino-Pearson):')
datasets = [dataset_Mittelwert_TR_5_Jahre, dataset_relative_Steigung_PAI, dataset_relative_Steigung_TR,
            dataset_Portfolio_Size, dataset_Patent_Asset_Index, dataset_Technology_Relevance,
           dataset_Competitive_Impact, dataset_Patentkennzahl, dataset_Rendite]
dataset_labels = ["technlogy relevance 5 years", "relative gradient Patent Asset Index", 'relative gradient technology relevance',
                 'Portfolio Size', 'Patent Asset Index', 'Technology Relevance year 2009', 'Competitive Impact',
                 'Patent Performance Index', 'Stock Return']

i=0
for dataset in datasets:
    k2, p = stats.normaltest(dataset)
    if i == 0:
        SL_p_value = p
    print('%10s %1.6f' % (dataset_labels[i],p))
    i += 1
#alpha = 0.001    
```
